```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4,5])
plt.show()
```

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25])
plt.show()
```

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25],'r^-')
plt.show()
# r - red color , o - for dots, - for line, ^ for triangles, s for square
```

```python
plt.title('Squares of values')
plt.xlabel('values',fontsize=18, color='red')
plt.ylabel('squares',fontsize=18, color='green')
plt.plot([1,2,3,4,5],[1,4,9,16,25],'bs-')
plt.show()
```

```python
# to change the axis values
plt.plot([1,2,3,4,5],[1,4,9,16,25],'bs-', linewidth=2.0)
plt.axis([0,10,0,30])
# for showing the grid
plt.grid(True)
plt.show()
```

```python
import numpy as np
x = np.array([0,5])
plt.plot(x,x**2,'r-',x,x**3,'g-',x,x**4,'y-')
# plt.grid(True)
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
x = [random.randint(1,100) for i in range(100)]
y = [random.randint(1,100) for i in range(100)]
plt.scatter(x,y)
plt.show()
a = [random.randint(1,100) for i in range(100)]
b = [random.randint(1,100) for i in range(100)]
plt.bar(x,y,a,b)
plt.bar(a,b)
plt.show()
```
